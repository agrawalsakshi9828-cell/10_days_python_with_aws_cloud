```python
student = {}
```

```python
def add_student (studnet_id , name , marks ):
    student[student_id] = {"name": name , "marks": marks }
    print(name , "Added Successfully")
```

```python
def view_all_students():
    if len(student) == 0:
        print("NO STUDENT FOUND")
        return 
    for student_id , details in student.items():
        print(student_id , " - ", details["name"] , "- Marks : ", details["marks"])
```

```python
def search_students(student_id):
    if student_id in student:
        details = student[student_id]
        print("FOUND", details["name"], "- Marks ",details["marks"] )
    else:
        print("NO Student found with id :" , student_id)
```

```python
def show_menu():
    print("""
    -----------------------------
   | 1. Add Students             |
   | 2. View All Students        |
   | 3. Search Students          |
   | 4. Exit                     |
    -----------------------------
    """)
```

```python
def run_program():
    while True:
        show_menu()
        choice = input("Enter Your CHoice :")

        if choice == "1":
            studnet_id  = input("Enter Student ID  :")
            name = input("Enter Student Name  :")
            marks = int(input("Enter Student Marks :"))
            add_student(student_id, name,marks)

        elif choice == "2":
            view_all_students()
        elif choice == "3":
            studnet_id  = input("Enter Student ID  :")
            search_students(student_id)
        elif choice == "4":
            break 
        else:
            print("ENter Valid Choice from 1 to 4")
            
```

```python
run_program()
```
